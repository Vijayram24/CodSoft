```python
import pandas as pd
```

```python
df = pd.read_csv("IMDb Movies India.csv"   ,encoding='cp1252')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe().T
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.Year = df.Year.str.replace('(' , '')
df.Year = df.Year.str.replace(')' , "")
df.Year = df.Year.astype('float64')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Duration'] = df.Duration.str.replace(' min' , '').astype('float64')
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.Year.fillna(method = 'ffill' , inplace = True)
```

```python
df.Duration.fillna(df.Duration.mean() , inplace = True)
```

```python
df.Genre.fillna(method = 'bfill' , inplace = True)
```

```python
df.Rating.unique()
```

```python
df.Rating.fillna(df.Rating.mean() , inplace = True)
```

```python
df.shape
```

```python
df.Votes.value_counts()
```

```python
df['Votes'] = df['Votes'].replace('[^\d.]', '', regex=True).astype('float64')
```

```python
df.Votes.fillna(df.Votes.mean() , inplace = True)
```

```python
df.isnull().sum()
```

```python
df.Director.unique()
```

```python
df.Director.fillna(df.Director.mode()[0] , inplace = True)
```

```python
act = ['Actor 1' , 'Actor 2' , 'Actor 3']
for i in act:
    df[i].fillna(df[i].mode()[0] , inplace = True)
```

```python
df.isnull().sum()
```

```python
li.append('Rating')
y = df['Rating']
x = df.drop(columns = li , axis = 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
model1.fit(x_train, y_train)
```

```python
y_pred = model1.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error , r2_score , accuracy_score
```

```python
mean_absolute_error(y_test , y_pred )
```
